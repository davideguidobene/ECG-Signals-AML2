```python
import pandas as pd
import numpy as np
```

```python
p_xgb_13 = 0.843509187908
p_rf = 0.822762299941
p_ensemble_stacking = 0.838174273859
p_xgb_7 = 0.844101956135
p_cat = 0.835210432721
```

```python
path = "to_ensemble/"
```

```python
df_xgb_13 = np.ravel(pd.read_csv(path + "output_xgb_13.csv", index_col=['id']))
df_rf = np.ravel(pd.read_csv(path + "output_random_forest_3.csv", index_col=['id']))
df_ensemble_stacking = np.ravel(pd.read_csv(path + "output_ensemble_stacking_2.csv", index_col=['id']))
df_xgb_7 = np.ravel(pd.read_csv(path + "output_xgb_7.csv", index_col=['id']))
df_cat = np.ravel(pd.read_csv(path + "output_cat_2.csv", index_col=['id']))
```

```python
l = []
for i in range(df_xgb_13.shape[0]):
    d = [0, 0, 0, 0]
    #d[df_xgb_13[i]] += p_xgb_13
    d[df_rf[i]] += p_rf
    #d[df_ensemble_stacking[i]] += p_ensemble_stacking
    d[df_xgb_7[i]] += p_xgb_7
    d[df_cat[i]] += p_cat
    print(d)
    l.append(np.argmax(d))
```

```python
l[-1]
```

```python
y_test = np.ravel(l)
```

```python
data_y_test = {
    'id': [i for i in range(len(y_test))],
    'y': y_test
}
df_y_test = pd.DataFrame(data_y_test)
df_y_test.to_csv('outputs/output_majority_voting.csv', index=False)
```
